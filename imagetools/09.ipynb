```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ng dn hnh nh
my_path = 'C:/Users/hieun/Downloads/hinhanh/hinhanhchieu.jpg'
#c nh
im = Image.open(my_path)
plt.imshow(im)
```

```python
#chuyn nh sang mu xm
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#to mt hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('biu  histogram ca hnh nh')
plt.xlabel('gi tr ca pixel')
plt.ylabel('s lng')
plt.show()
```

```python
#lm ni bt ng bin ca nh
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
