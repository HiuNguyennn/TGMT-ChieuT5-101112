```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#đường dẫn hình ảnh
my_path = 'C:/Users/hieun/Downloads/hinhanh/hinhanhchieu.jpg'
#đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
#chuyển ảnh sang màu xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('biểu đồ histogram của hình ảnh')
plt.xlabel('giá trị của pixel')
plt.ylabel('số lượng')
plt.show()
```

```python
#làm nổi bật đường biên của ảnh
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
#đường dẫn ảnh
my_path = 'C:/Users/hieun/Downloads/hinhanh/hinhanhdomat.jpg'
#đọc và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")
#hiển thị ảnh trên trục tọa độ
plt.imshow(im)
#tạo một hình vẽ mới
plt.figure()
#đặt màu cho ảnh là xám
plt.gray()
#vẽ biểu đồ đường biên của ảnh
plt.contour(im, origin="image")
#đảm bảo tỉ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')
#hiển thị ảnh
plt.show()
```
