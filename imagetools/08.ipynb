```python
import matplotlib.pyplot as plt
```

```python
# V biu : Line Chart
x = [1, 2, 3, 4, 5]
y = [20,14 , 10, 22, 15]

x1 = [1, 2, 3, 4, 5]
y1 = [22, 13, 18, 9, 20]

plt.plot(x, y, 'yo-', label='D liu') # V ng 1
plt.plot(x1, y1, marker='o', color='k', linestyle='-', label='D liu 2')# V ng 2
plt.legend() # Hin th ch gii
plt.xlabel('Trc x') # t nhn cho trc x
plt.ylabel('Trc y') # t nhn cho trc y
plt.show() # Hin th biu 
```

```python
# V biu  ct Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 10, 30, 15, 25]

plt.bar(categories, values, color='y', alpha=0.6) # V biu  ct
plt.xlabel('Danh mc') # t nhn cho trc x
plt.ylabel('Gi tr') # t nhn cho trc y
plt.show() # Hin th biu 
```

```python
# V biu  hnh trn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [25, 10, 35, 30]
my_colors = ['blue', 'green', 'red', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # V biu  hnh trn
plt.title('Biu  hnh trn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [18, 22, 11, 27, 14]

plt.scatter(x, y, c='k', marker='o', label='D liu')  # V biu  phn tn
plt.title('Biu  phn tn v d')  # t tiu  cho biu 
plt.xlabel('Trc x')  # t nhn cho trc x
plt.ylabel('Trc y')  # t nhn cho trc y
plt.legend()  # Hin th ch gii
plt.show()  # Hin th biu 
```
