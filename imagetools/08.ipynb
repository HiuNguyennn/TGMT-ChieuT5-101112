```python
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ: Line Chart
x = [1, 2, 3, 4, 5]
y = [20,14 , 10, 22, 15]

x1 = [1, 2, 3, 4, 5]
y1 = [22, 13, 18, 9, 20]

plt.plot(x, y, 'yo-', label='Dữ liệu') # Vẽ đường 1
plt.plot(x1, y1, marker='o', color='k', linestyle='-', label='Dữ liệu 2')# Vẽ đường 2
plt.legend() # Hiển thị chú giải
plt.xlabel('Trục x') # Đặt nhãn cho trục x
plt.ylabel('Trục y') # Đặt nhãn cho trục y
plt.show() # Hiển thị biểu đồ
```

```python
# Vẽ biểu đồ cột Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 10, 30, 15, 25]

plt.bar(categories, values, color='y', alpha=0.6) # Vẽ biểu đồ cột
plt.xlabel('Danh mục') # Đặt nhãn cho trục x
plt.ylabel('Giá trị') # Đặt nhãn cho trục y
plt.show() # Hiển thị biểu đồ
```

```python
# Vẽ biểu đồ hình tròn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [25, 10, 35, 30]
my_colors = ['blue', 'green', 'red', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [18, 22, 11, 27, 14]

plt.scatter(x, y, c='k', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [24, 16, 9, 22, 31, 14, 53, 31, ]

plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [10, 19, 24, 28, 33, 41, 56, 59, 64]

plt.violinplot(data)  # Vẽ biểu đồ violin
plt.title('Biểu đồ violin ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```
